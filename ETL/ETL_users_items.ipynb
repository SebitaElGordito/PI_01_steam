## Se realiza proceso de ETL sobre dataset users_items

Importamos las librerías

```python
import pandas as pd
```

<br>
<br>

Se extraen los datos del dataset provisto por el cliente

```python
df_i = pd.read_csv(r'C:\Users\Cebol\OneDrive\Escritorio\Datasets\Datasets\australian_users_items.csv', encoding='utf-8')

df_i
```

<br>
<br>

Se observan los tipos de datos y cantidad de nulos

```python
dicc = {"Campo": [], "Tipo_de_dato": [], "Nulos": []}

for columna in df_i.columns:
    dicc["Campo"].append(columna)
    dicc["Tipo_de_dato"].append(df_i[columna].apply(type).unique())
    dicc["Nulos"].append(df_i[columna].isnull().sum())

df_info = pd.DataFrame(dicc)
df_info
```

<br>
<br>

Se normalizan los valores de la columna 'item_name', llevándolos a todos los registros a minúscula

```python
df_i['item_name'] = df_i['item_name'].str.lower() 
```

<br>
<br>

Se verifican los duplicados en el Dataframe, filtrando por algunos subsets, dado que llamando a la función únicamente, no obtengo ningún resultado

```python
filas_duplicadas = df_i[df_i.duplicated(subset=["user_id", "steam_id", "item_id", "playtime_forever"], keep=False)]
filas_duplicadas
```

<br>
<br>

Se decide eliminar los registros duplicados

```python
df_i = df_i.drop_duplicates(subset=["user_id", "steam_id", "item_id", "playtime_forever"], keep='first')
df_i.shape
```

<br>
<br>

Se chequea la cantidad de valores nulos por columna

```python
df_i.isna().sum()
```

<br>
<br>

Se procede a la eliminación de los registros que no tienen datos nulos de los item_id, dado que no nos sirven, y seguramente se lleven consigo los nulos en las demas columnas.

```python
df_i = df_i.dropna(subset=["item_id"])
df_i.isna().sum()
```

<br>
<br>

Se procede a la eliminación de las columnas que no se van a utilizar en la API, ni en nuestro modelo de recomendación. Se borra la columna 'Unnamed:0', la columna 'playtime_2weeks', dado que solo utilizaremos la columna que nos muestra el tiempo total jugado, la columna 'user_url', y la columna 'items' dado que es la columna que contiene el resto de columnas que fueron desanidadas. Por ultimo se decide eliminar el campo 'steam_id' dado que no le encuentro relación con ninguna tabla, y no va a ser utilizado el dato.

```python
df_i = df_i.drop(['Unnamed: 0', 'playtime_2weeks', 'user_url', 'items', 'steam_id' ], axis=1)
df_i.columns
```

<br>
<br>

El dataset quedaría conformado de la siguiente manera

```python
df_i
```

```python
dicc = {"Campo": [], "Tipo_de_dato": [], "Nulos": []}

for columna in df_i.columns:
    dicc["Campo"].append(columna)
    dicc["Tipo_de_dato"].append(df_i[columna].apply(type).unique())
    dicc["Nulos"].append(df_i[columna].isnull().sum())

df_info = pd.DataFrame(dicc)
df_info
```
